# Weather prediction For Eindhoven

## Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the CSV data into a DataFrame


```python
data = pd.read_csv('eindhoven-weather.csv', parse_dates=['time'])
```

## Describe basic info about the dataset

```python
print(data.head())

print(data.info())

print(data.isnull().sum())

print(data.describe())
```

## The dataset does not have missing values, but we will add the following line just for exercise


```python
data.fillna(0, inplace=True)
```

## Remove duplicates

```python

data.drop_duplicates(inplace=True)
```

## Plot temperature over time

```python
plt.figure(figsize=(12, 6))
plt.plot(data['time'], data['temperature_2m (째C)'], label='Temperature (째C)')
plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (째C)')
plt.legend()
plt.show()
```

## Correlation heatmap


```python

plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
