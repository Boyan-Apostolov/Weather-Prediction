# Weather prediction For Eindhoven

## Preparing the data

### Data Dictionary

```python
data_dictionary = {
    "time": "Timestamp of the recorded weather data (datetime)",
    "temperature_2m (°C)": "Temperature at 2 meters above ground (float, °C)",
    "relative_humidity_2m (%)": "Relative humidity at 2 meters above ground (float, %)",
    "precipitation (mm)": "Total precipitation recorded (float, mm)",
    "rain (mm)": "Rainfall recorded (float, mm)",
    "pressure_msl (hPa)": "Atmospheric pressure at mean sea level (float, hPa)",
    "wind_speed_10m (km/h)": "Wind speed at 10 meters above ground (float, km/h)"
}

print("\nData Dictionary:")
for key, value in data_dictionary.items():
    print(f"{key}: {value}")
```

### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the CSV data into a DataFrame


```python
data = pd.read_csv('eindhoven-weather.csv', parse_dates=['time'])
```

### Describe basic info about the dataset

```python
print(data.head())

print(data.info())

print(data.isnull().sum())

print(data.describe())
```

### The dataset does not have missing values, but we will add the following line just for exercise


```python
data.fillna(0, inplace=True)
```

### Remove duplicates

```python
data.drop_duplicates(inplace=True)
```

## Data visualization 

### Plot temperature over time

```python
plt.figure(figsize=(12, 6))
plt.plot(data['time'], data['temperature_2m (°C)'], label='Temperature (°C)')
plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

- The graph shows a clear seasonal temperature cycle from 2020 to 2025, with peaks in warmer months and dips in colder months. The pattern remains consistent, indicating stable annual fluctuations with occasional extreme variations

### Correlation heatmap


```python

plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

The correlation heatmap shows relationships between different weather variables. Key observations:
- Precipitation and Rain are highly correlated (0.99), indicating they often occur together.
- Temperature and Humidity have a moderate negative correlation (-0.54), meaning higher temperatures are associated with lower humidity.
- Pressure, Wind Speed, and Precipitation show weak correlations with other variables, suggesting minimal direct influence.
